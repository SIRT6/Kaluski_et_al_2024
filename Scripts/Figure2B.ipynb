```r
library(ggplot2)
library(dplyr)
library(ggrepel)
library(DOSE)
library(wesanderson)
library(ggpubr)
library(MASS)
```

```r
fko <- read.csv("../Data/FullKO_lfc.csv")
bko <- read.csv("../Data/brKO_lfc.csv")
```

```r
df <- inner_join(fko, bko, by = c("X"="X"))
```

```r
colnames(df) <- c("Gene", "Full", "Brain")
```

```r
df$Mean <- rowMeans(df[,c(2,3)])
```

```r
ggplot(df, aes(x = Full, y = Brain, label = Gene)) +
    geom_rect(aes(xmin = 0, xmax = Inf, 
                  ymin = 0, ymax = Inf), 
                  #fill = "#9ec4e7",
                  fill = wes_palette("Moonrise3")[4], 
                  alpha = 0.05) +
    geom_hline(yintercept = 0, linetype = "dashed", color = "#393433", linewidth = 0.5) +
    geom_vline(xintercept = 0, linetype = "dashed", color = "#393433", linewidth = 0.5) + 
    geom_point(color = "black", size = 5.8, alpha = 1) +
    geom_point(color = "#588ebb", size = 5, alpha = 1) + 
    coord_cartesian(clip = 'off')+
    scale_y_continuous(name = expression(log[2]("FC")~"brS6KO vs WT"), 
                     limits = c(-1, 6), breaks = seq(-1, 6, 1)) + 
    scale_x_continuous(name = expression(log[2]("FC")~"SIRT6 KO vs WT"), 
                     limits = c(-0.25, 0.5)) + 
    theme_dose(21)
```
